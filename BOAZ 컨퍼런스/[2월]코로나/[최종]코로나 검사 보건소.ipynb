```python
import time
from bs4 import BeautifulSoup
from urllib.request import urlopen
import pandas as pd
import requests
```

```python
url = 'http://www.mohw.go.kr/react/popup_200128.html'
r = requests.get(url)
soup = BeautifulSoup(r.content, 'html.parser', from_encoding = 'utf-8')
```

```python
soup
```

```python
시도 = []
시군구 = []
선별진료소 = []
전화번호 = []
```

```python
td = soup.find_all('td')
```

```python
for t in range(len(td)):
    if t % 4 == 0:
        시도.append(td[t].text)
    elif t % 4 == 1:
        시군구.append(td[t].text)
    elif t % 4 == 2:
        선별진료소.append(td[t].text)
    else:
        전화번호.append(td[t].text)
```

```python
전화번호
```

```python
df = pd.DataFrame({'시/도' : 시도, '시군구' : 시군구, '선별진료소' : 선별진료소, '전화번호' : 전화번호})
```

```python
df
```

```python
for i in range(len(df['선별진료소'])):
    if '검체채취 가능' in df['선별진료소'][i]:
        df['선별진료소'][i] = df['선별진료소'][i].split('*')[0][:-1]
    else:
        pass
```
