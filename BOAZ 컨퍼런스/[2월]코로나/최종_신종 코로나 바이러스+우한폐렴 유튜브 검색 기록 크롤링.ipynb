```python
def youtube_url_crawling(search_word):
    # search_word 검색
    driver.find_element_by_xpath('//*[@id="search"]').send_keys(search_word)
    driver.find_element_by_xpath('//*[@id="search-icon-legacy"]').click()
    time.sleep(2)
    
    # 채널, 재생목록 제외하고 동영상만 검색되도록 필터링
    driver.find_element_by_xpath('//*[@id="container"]/ytd-toggle-button-renderer/a').click()
    time.sleep(2)
    driver.find_element_by_xpath('//*[@id="label"]/yt-formatted-string').click()
```

```python
def store_url():
    url_href = []
    no_url = []
    i = 1
    try:
        while True:
            href = driver.find_element_by_xpath('/html/body/ytd-app/div/ytd-page-manager/ytd-search/div[1]/ytd-two-column-search-results-renderer/div/ytd-section-list-renderer/div[2]/ytd-item-section-renderer/div[3]/ytd-video-renderer[' + str(i) + ']/div[1]/ytd-thumbnail/a').get_attribute('href')
            url_href.append(href)
            i += 1
    except:
        no_url.append(i)
        pass
    print(len(url_href))
    return url_href
```

## 크롤링

```python
titles = []            # 제목
dates = []             # 영상 업로드 날짜
hits = []             # 조회수
descriptions = []    #상세설명
```

```python
def make_youtube_dataframe (url_list):
    for i in url_list:
        driver.get(i)
        page = driver.page_source
        soup = BeautifulSoup(page, 'lxml')
        
        # 제목 크롤링
        try:
            title = soup.find_all('yt-formatted-string', {'class' : 'style-scope ytd-video-primary-info-renderer'})[0].text
            titles.append(title)
        except:
            titles.append(None)
        
        # 조회수 크롤링
        try:
            hit = soup.find_all('span', {'class' : 'view-count style-scope yt-view-count-renderer'})[0].text
            hits.append(hit)
        except:
            hits.append(None)
        
        # 날짜 크롤링
        try:
            date = soup.find_all('yt-formatted-string', {'class' : 'style-scope ytd-video-primary-info-renderer'})[1].text
            dates.append(date)
        except:
            dates.append(None)

        time.sleep(2)
    
    # 데이터프레임 만들기
    df = pd.DataFrame({'title' : titles, 'date' : dates, 'hit' : hits, 'url' : url_list})
    return df
```

```python
youtube = make_youtube_dataframe(no_repeat_total_url)
```

## date column 전처리

1. 문자 없애기(ex. 실시간 스트리밍, 최초 시작일 등등)  
2. '.'을 '-'로 바꿔서 date 타입으로 변환

```python
import re
from datetime import datetime
for i in range(len(cp_youtube['date'])):
    try:
        data = cp_youtube['date'][i]
    
        step1 = re.sub('[^\d*\.\d]', '', data)
        step2 = step1[0:-1].replace('.', '-')
    
        step3 = datetime.strptime(step2, '%Y-%m-%d')
    
        step4 = str(step3)
        cp_youtube['date'][i] = step4[0:10]
    
    except:
        pass
```
